```python
import pandas as pd
from imblearn.over_sampling import SMOTE 
```

```python
df = pd.read_csv('./Test_02.csv')
```

```python
df[:3]
```

```python
target = 'Potability'
X, y = df.drop(target, axis=1), df[target]

sampling_df = None # sampling_df 초기화
X_over_resampled, y_over_resampled = SMOTE().fit_resample(X, y)          # sampling_strategy 옵션: 'auto', sampling_strategy
sampling_df = X_over_resampled
sampling_df[target] = y_over_resampled
```

```python
print(len(df))
print(len(sampling_df))
```

```python
df_before = df
df_after = sampling_df

# df_before['state'] = 'Before'
# df_after['state'] = 'After'
```

```python
column = df_before.columns
test = []
for idx, col in enumerate(df_before.columns):
        combined_df = pd.DataFrame({
            col: df_before[col],
            'sampled_' + col : df_after[col]
        })
        test.append(combined_df)
```
